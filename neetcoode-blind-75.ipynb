```python
# 1 Two sum
# https://leetcode.com/problems/two-sum/
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hm = {}
        for i, v in enumerate(nums):
            diff = target - v
            if diff in hm:
                return [hm[diff], i]
            hm[v] = i
        return -1

s = Solution()
assert s.twoSum(nums = [2,7,11,15], target = 18) == [1, 2]
```

```python
# 238. Product of Array Except Self
# https://leetcode.com/problems/product-of-array-except-self/
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        ln = len(nums)
        ans = [1] * ln

        prefix = 1
        for i in range(ln):
            ans[i] *= prefix
            prefix *= nums[i]

        suffix = 1
        for i in range(ln - 1, -1, -1):
            ans[i] *= suffix
            suffix *= nums[i]

        return ans


s = Solution()
s.productExceptSelf(nums = [1,2,3,4]) # [24,12,8,6]
```

```python
# 15. 3Sum
# https://leetcode.com/problems/3sum/
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:

        ans = []
        nums.sort()

        for i, v in enumerate(nums):
            if i > 0 and v == nums[i - 1]:
                continue
            l, r = i + 1, len(nums) - 1
            while l < r:
                sum_3 = v + nums[l] + nums[r]
                if sum_3 > 0:
                    r -= 1
                elif sum_3 < 0:
                    l += 1
                else:
                    ans.append([v, nums[l], nums[r]])
                    l += 1
                    while l < r and nums[l] == nums[l - 1]:  # l < r before nums[l] == nums[l - 1]
                        l += 1
        return ans
    
# Line 10, 11 and 22, 23 are all about handling duplicates

    
s = Solution()
assert s.threeSum(nums = [-1,0,1,2,-1,-4]) == [[-1, -1, 2], [-1, 0, 1]]
```

```python
# 11. Container With Most Water
# https://leetcode.com/problems/container-with-most-water/
class Solution:
    def maxArea(self, height: list[int]) -> int:
        l, r, mx = 0, len(height) - 1, 0
        while l < r:
            mx = max(mx, (r - l) * min(height[l], height[r]))
            if height[l] < height[r]:
                l += 1
            else:
                r -= 1
        return mx

s = Solution()
assert s.maxArea(height = [1,8,6,2,5,4,8,3,7]) == 49
```

```python
# 424. Longest Repeating Character Replacement
# https://leetcode.com/problems/longest-repeating-character-replacement/
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        """  
        Key insights: The window is valid if the window length minus the most frequent 
        character's count is less than or equal to the number of allowed replacements.
        """
        l, mx, c = 0, 0, {}
        most_count = 0
        for r in range(len(s)):
            c[s[r]] = c.get(s[r], 0) + 1
            most_count = max(most_count, c[s[r]])
            if r - l + 1 - most_count > k:
                c[s[l]] -= 1
                l += 1
        return r - l + 1


solution = Solution()
solution.characterReplacement(s = "ABAB", k = 2)
```

```python
# 20. Valid Parentheses
# https://leetcode.com/problems/valid-parentheses/
from collections import deque
class Solution:
    def isValid(self, s: str) -> bool:
        hm = {
            ')': '(',
            ']': '[',
            '}': '{'
        }

        stack = []
        for c in s:
            if stack and stack[-1] == hm.get(c):
                stack.pop()
            else:
                stack.append(c)
        return not stack

solution = Solution()
assert solution.isValid(s = "()[]{}") is True
assert solution.isValid(s = "(]") is False
assert solution.isValid(']') is False
```

```python
# 206. Reverse Linked List
# https://leetcode.com/problems/reverse-linked-list/
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def graph(self):
        nodes = []
        while self:
            nodes.append(str(self.val) + ' ->')
            self = self.next
        return ' '.join(nodes) + ' None'
    
def list2ll(arr):
    crt = dummy = ListNode()
    for a in arr:
        crt.next = ListNode(a)
        crt = crt.next
    return dummy.next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev, crt = None, head
        while crt:
            nxt = crt.next
            crt.next = prev
            prev = crt
            crt = nxt
        return prev

s = Solution()
new_head = s.reverseList(list2ll([1,2,3,4,5]))
assert new_head.graph() == '5 -> 4 -> 3 -> 2 -> 1 -> None'
```

```python
# 21. Merge Two Sorted Lists
# https://leetcode.com/problems/merge-two-sorted-lists/
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        crt = dummy

        l, r = list1, list2 

        while l and r:
            if l.val < r.val:
                crt.next = l
                l = l.next

            else:
                crt.next = r
                r = r.next
            crt = crt.next

        if l:
            crt.next = l

        if r:
            crt.next = r

        return dummy.next


list1 = list2ll([1,2,4])
list2 = list2ll([1,3,4])
s = Solution()
head = s.mergeTwoLists(list1, list2)
assert head.graph() == '1 -> 1 -> 2 -> 3 -> 4 -> 4 -> None'
```

```python
# Reorder List
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """

        # find middle
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # reverse the 2nd half
        crt = slow.next
        slow.next = None # set the first half's last.next to None
        prev = None
        while crt:
            nxt = crt.next
            crt.next = prev
            prev = crt
            crt = nxt

        # merge 2 halves
        first, second = head, prev
        dummy = ListNode()
        crt = dummy
        while first and second:
            crt.next = first
            first = first.next
            crt = crt.next

            crt.next = second
            second = second.next
            crt = crt.next
        
        if first:
            crt.next = first

        return dummy.next
```

```python
# Remove Nth Node From End of List
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(next=head)
        l, r = dummy, head
        while n > 0:
            r = r.next
            n -= 1

        while r:
            l = l.next
            r = r.next

        # delete
        l.next = l.next.next
        return dummy.next
```

```python
# Linked List Cycle
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        s, f = head, head
        while f and f.next:
            s = s.next
            f = f.next.next
            if s == f:
                return True
        return False
```

```python
# 104. Maximum Depth of Binary Tree
# https://leetcode.com/problems/maximum-depth-of-binary-tree/
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        depth = 0
        q = deque()
        if root:
            q.append(root)
            while q:
                level_size = len(q)
                depth += 1
                
                for _ in range(level_size):
                    node = q.popleft()
                    if node.left:
                        q.append(node.left)
                    if node.right:
                        q.append(node.right)
        return depth
    
s = Solution()
root = deserialize([3,9,20,None,None,15,7,None,None,None,None])
root.graph()
assert s.maxDepth(root) == 3
```

```python
# 102. Binary Tree Level Order Traversal
# https://leetcode.com/problems/binary-tree-level-order-traversal/
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        q = deque()
        if root:
            q.append(root)

        ans = []
        while q:
            lvl = []
            for _ in range(len(q)):
                crt = q.popleft()
                lvl.append(crt.val)

                if crt.left:
                    q.append(crt.left)
                if crt.right:
                    q.append(crt.right)
            ans.append(lvl)
        return ans

root = deserialize([3,9,20,None,None,15,7,None,None,None,None])
root.graph()
s = Solution()
s.levelOrder(root)
```

```python
# Kth Smallest Element In a Bst
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        self.idx = 0
        self.ans = None
        def inorder(root):  
            if not root:
                return
            inorder(root.left)
            val = root.val
            self.idx += 1
            if self.idx == k:
                self.ans = val
                return
            inorder(root.right)
        inorder(root)
        return self.ans
```

```python
# Binary Tree Maximum Path Sum
# https://www.youtube.com/watch?v=mOdetMWwtoI
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        self.mx = float('-inf')
        def pathSum(root):
            if root is None:
                return 0
            left = max(0, pathSum(root.left)) # if a node val is negative, we can simply add 0 (don't add) 
            right = max(0, pathSum(root.right)) # since it won't contribute to max sum
            self.mx = max(self.mx, left + right + root.val)
            return max(left, right) + root.val
        pathSum(root)
        return self.mx
```

```python
# Serialize And Deserialize Binary Tree
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        def preorder(root):
            if root is None:
                return ['1001']
            left = preorder(root.left)
            right = preorder(root.right)
            return [str(root.val), *left, *right]
        return ', '.join(preorder(root))



    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        data = data.split(', ')
        self.idx = 0
        def preorder():
            if data[self.idx] == '1001':
                self.idx += 1
                return None
            root = TreeNode(int(data[self.idx]))
            self.idx += 1
            root.left = preorder()
            root.right = preorder()
            return root
        return preorder()
```

```python
# Design Add And Search Words Data Structure
class WordDictionary:

    def __init__(self):
        self.d = {}
        

    def addWord(self, word: str) -> None:
        crt = self.d
        for c in word:
            if c not in crt:
                crt[c] = {}
            crt = crt[c]
        crt['*'] = {}
        

    def search(self, word: str) -> bool:
        def dfs(idx, crt):
            if idx == len(word):
                return '*' in crt

            if word[idx] == '.':
                for _, v in crt.items():
                    if dfs(idx + 1, v):
                        return True
                return False

            return word[idx] in crt and dfs(idx + 1, crt[word[idx]])

        return dfs(0, self.d)

```

```python
# Find Median From Data Stream
# https://www.youtube.com/watch?v=JNKdfHmnMSg
from heapq import heappush, heappop
class MedianFinder:

    def __init__(self):
        self.mn = []
        self.mx = []
        self.size = 0
        

    def addNum(self, num: int) -> None:
        self.size += 1
        if not self.mx or num < -self.mx[0]:
            heappush(self.mx, -num)
        else:
            heappush(self.mn, num)

        if len(self.mx) > len(self.mn) + 1:
            heappush(self.mn, -heappop(self.mx))
        elif len(self.mx) < len(self.mn):
            heappush(self.mx, -heappop(self.mn))        

    def findMedian(self) -> float:
        if self.size & 1 != 0:
            return -self.mx[0]
        return (-self.mx[0] + self.mn[0]) / 2
    

obj = MedianFinder()
obj.addNum(1)
obj.addNum(2)
print(obj.findMedian())
obj.addNum(3)
print(obj.findMedian())
```

```python
# Combination Sum
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        ans = []
        def dfs(i, crt, vals): # i == idx of candidates, crt == crt_sum,  vals = combinations
            if crt == target:
                ans.append([*vals])
                return
            
            if i == len(candidates) or crt > target:
                return
            vals.append(candidates[i])
            dfs(i, crt + candidates[i], vals)
            vals.pop()

            dfs(i + 1, crt, vals)

        dfs(0, 0, [])
        return ans
```

```python
# Clone Graph
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

from collections import deque
class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        # we can use Node.val as key in hashmap because Node.val is unique for each node constraint.
        if not node:
            return None

        visited = {node.val: Node(node.val)}
        q = deque([node])

        while q:
            crt = q.popleft()
            crt_cp = visited[crt.val]
            
            for nbr in crt.neighbors:
                if nbr.val not in visited:
                    visited[nbr.val] = Node(nbr.val)
                    q.append(nbr)
                crt_cp.neighbors.append(visited[nbr.val])
        return visited[node.val]
```

```python
# Graph Valid Tree
# https://www.lintcode.com/problem/178/
# https://www.youtube.com/watch?v=gr5ccsYJ3l4
# It's about detecting cycle in an undirected graph
from collections import defaultdict
class Solution:
    """
    @param n: An integer
    @param edges: a list of undirected edges
    @return: true if it's a valid tree, or false
    """
    def valid_tree(self, n: int, edges: List[List[int]]) -> bool:
        graph = defaultdict(list)
        for e1, e2 in edges:
            graph[e1].append(e2)
            graph[e2].append(e1)
        
        visited = set()
        def has_cycle(child, parent):
            if child in visited:
                return True
            visited.add(child)

            for nbr in graph[child]:
                if nbr == parent:
                    continue
                if has_cycle(nbr, child):
                    return True
            return False

        # if has_cycle(0, -1): 
        #     return False
        # return n == len(visited)
        return not has_cycle(0, -1) and n == len(visited)



n, edges = 5,  [[0, 1], [0, 2], [0, 3], [1, 4]]
# n, edges = 5, [[0, 1], [1, 2], [2, 3], [1, 3], [1, 4]]
sln = Solution()
sln.valid_tree(n, edges)
```

```python
# Alien Dictionary
# https://www.youtube.com/watch?v=IIgisEKjCKA
class Solution:
    """
    @param words: a list of words
    @return: a string which is correct order
    """
    def alien_order(self, words: List[str]) -> str:

        graph = {}
        for word in words:
            for c in word:
                graph[c] = []

        for i in range(len(words) - 1):
            l, r = words[i], words[i + 1]
            lln, rln = len(l), len(r)
            mn = min(lln, rln)
            if lln > rln and l[:mn] == r[:mn]:
                return ''
            for j in range(mn):
                if l[j] != r[j]:
                    graph[l[j]].append(r[j])
                    break

        
        ans, visited = [], {}
        def has_cycle(c):
            if c in visited:
                return visited[c]
            visited[c] = True
            
            for nbr in graph[c]:
                if has_cycle(nbr):
                    return True
            visited[c] = False
            ans.append(c)
            return False

        # for c in graph: It's ok for leetcode here but for lintcode 'return the smallest in normal lexicographical order' we need the below code
        for c in sorted(graph.keys(), reverse=True):
            if has_cycle(c):
                return ''
        ans.reverse()
        return ''.join(ans)

words = ["wrt","wrf","er","ett","rftt"]
# words = ["wrth","wrt","wrf","er","ett","rftt"]
sln = Solution()
sln.alien_order(words)
```

```python
# 198. House Robber
# https://leetcode.com/problems/house-robber/

class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        if n > 1:
            nums[1] = max(nums[0], nums[1])
        for i in range(2, n):
            nums[i] = max(nums[i] + nums[i - 2], nums[i - 1])
        return nums[-1]

s = Solution()
assert s.rob(nums = [2,7,9,3,1]) == 12
```

```python
# 213. House Robber II
class Solution:
    def rob(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        def helper(nums):
            n = len(nums)
            if n > 1:
                nums[1] = max(nums[0], nums[1])
            for i in range(2, n):
                nums[i] = max(nums[i] + nums[i - 2], nums[i - 1])
            return nums[-1]
        return max(helper(nums[:-1]), helper(nums[1:]))
```

```python
# 91. Decode Ways
class Solution:
    def numDecodings(self, s: str) -> int:
        if s[0] == '0':
            return 0

        n = len(s) + 1
        m = [0] * n
        m[0], m[1] = 1, 1

        for i in range(2, n):
            if s[i - 1] != '0':
                m[i] += m[i - 1]
            if '10' <= s[i - 2: i] <= '26':
                m[i] += m[i - 2]
        return m[-1]

s = '226' # [1, 1, 0, 0]
sln = Solution()
sln.numDecodings(s)
```

```python
# 322. Coin Change
# Trivial Example [1, 2, 5], 7 # 2 
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        n = amount + 1
        m = [n] * n
        m[0] = 0
        
        for i in range(1, n):
            for c in coins:
                if i >= c:
                    m[i] = min(m[i], m[i - c] + 1)
        return m[-1] if m[-1] != n else -1
    

sln = Solution()
sln.coinChange([1, 2, 5], 5) # 1
sln.coinChange([1, 2, 5], 7) # 2
```

```python
# 6 Maximum Product Subarray
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        mx = float('-inf')
        prev_mn, prev_mx = 1, 1

        for n in nums:
            crt_mn = min(n, n * prev_mn, n * prev_mx)
            crt_mx = max(n, n * prev_mn, n * prev_mx)

            mx = max(mx, crt_mx)

            prev_mn = crt_mn
            prev_mx = crt_mx
        return mx

nums = [-2, 2, 2]
nums = [-2, 2, 2, -2]
sln = Solution()
sln.maxProduct(nums)
```

```python
# Longest Increasing Subsequence
# https://leetcode.com/problems/longest-increasing-subsequence/
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        m = [1] * n
        for r in range(1, n):
            for l in range(r):
                if nums[l] < nums[r]:
                    m[r] = max(m[r], m[l] + 1)
        return max(m)

s = Solution()
s.lengthOfLIS([10,9,2,5,3,7,101,18]) # 4
# s.lengthOfLIS([0,1,0,3,2,3]) # 4
```

```python
# Jump Game
# https://www.youtube.com/watch?v=WpC158p7DAY
class Solution:
    def canJump(self, nums: List[int]) -> bool: 
        n = len(nums)
        ptr = n - 1
        for i in range(n - 2, -1, -1):
            if i + nums[i] >= ptr:
                ptr = i
        return ptr == 0
```

```python
# 56. Merge Intervals
# https://leetcode.com/problems/merge-intervals/
class Solution:
    def merge(self, itvls: List[List[int]]) -> List[List[int]]:
        itvls.sort(key=lambda x: x[0]) 
        # intervals.sort() != intervals.sort(key=lambda x: x[0])
        # intervals.sort() will sort by the first element first, and then by the second element
        ans = [itvls[0]]
        for start, end in itvls[1:]:
            last = ans[-1]
            if last[1] >= start:
                last[1] = max(last[1], end)
            else:
                ans.append([start, end])
        return ans

s = Solution()
assert s.merge([[1,3],[2,6],[8,10],[15,18]]) == [[1,6],[8,10],[15,18]]
```

```python
# 435. Non-overlapping Intervals
# https://leetcode.com/problems/non-overlapping-intervals/
class Solution:
    def eraseOverlapIntervals(self, itvls: List[List[int]]) -> int:
        itvls.sort(key=lambda x: x[0])
        last_end = itvls[0][1]
        ans = 0
        for start, end in itvls[1:]:
            if start < last_end:
                last_end = min(end, last_end)
                ans += 1
            else:
                last_end = end
        return ans

s = Solution()
# s.eraseOverlapIntervals(intervals = [[1,2],[2,3],[3,4],[1,3]])
# s.eraseOverlapIntervals(intervals = [[1,2],[1,2],[1,2]]) # 2
# s.eraseOverlapIntervals([[1,100],[11,22],[1,11],[2,12]]) # 2
# s.eraseOverlapIntervals([[0,2],[1,3],[2,4],[3,5],[4,6]]) # 2
assert s.eraseOverlapIntervals([[-52,31],[-73,-26],[82,97],[-65,-11],[-62,-49],[95,99],[58,95],[-31,49],[66,98],[-63,2],[30,47],[-40,-26]]) == 7
```

```python
# Meeting Rooms II
# https://leetcode.com/problems/meeting-rooms-ii/
class Solution:
    def min_meeting_rooms(self, intervals: List[List]) -> int:
        times = []
        for interval in intervals:
            times.append((interval[0], 1))
            times.append((interval[1], -1))

        times.sort(key=lambda x: x[0])
        mx, cnt = 0, 0
        for t in times:
            cnt += t[1]
            # print(cnt, t)
            mx = max(mx, cnt)
        return mx

s = Solution()
s.min_meeting_rooms([(0,30),(5,10),(15,20)]) # 2
# s.min_meeting_rooms([(2,7)]) # 1
```

```python
# 73. Set Matrix Zeroes
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """

        zeros = []
        rows, cols = len(matrix), len(matrix[0])
        for r in range(rows):
            for c in range(cols):
                if matrix[r][c] == 0:
                    zeros.append((r, c))

        # remove repeated work with set comprehension
        zero_rows = {r for (r, _) in zeros}
        zero_cols = {c for (_, c) in zeros}

        for r in range(rows):
            for c in zero_cols:
                matrix[r][c] = 0

        for r in zero_rows:
            for c in range(cols):
                matrix[r][c] = 0

matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
s = Solution()
s.setZeroes(matrix)
print(matrix == [[0, 0, 0, 0], [0, 4, 5, 0], [0, 3, 1, 0]])
```

```python
# Counting Bits
class Solution:
    def countBits(self, n: int) -> List[int]:
        t = [0] * (n + 1)
        bit = 1
        for i in range(1, n + 1):
            if bit * 2 == i:
                bit = i
            t[i] = 1 + t[i - bit]
        return t

n = 10
sln = Solution()
sln.countBits(n) 
```

```python
# Sum of Two Integers
class Solution:
    def getSum(self, a: int, b: int) -> int:
        bit_shortener = 2 ** 32 - 1

        while b & bit_shortener:
            carry = (a & b) << 1
            a ^= b
            b = carry
        return a & bit_shortener if b else a
    
a, b = 3, 2  # 11 10  (11 & 10) << 1 = 100 (carry) =>  11 ^ 10 = 01 (a) => b == 100 => 
#                     (001 & 100) << 1 = 0 (carry) =>  001 ^ 100 = 101(a) => b == 0 => end loop
sln = Solution()
sln.getSum(a, b)
```
