```python
class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        31. Next Permutation
        Do not return anything, modify nums in-place instead.
        """

        ln = len(nums)
        pivot = ln - 1
        while pivot > 0 and nums[pivot] <= nums[pivot - 1]:
            pivot -= 1

        l, r = pivot, ln - 1
        while l < r:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
            r -= 1
        if pivot == 0:
            return
        
        pivot -= 1
        for i in range(pivot + 1, ln):
            if nums[i] > nums[pivot]:
                nums[i], nums[pivot] = nums[pivot], nums[i]
                return

s = Solution()

nums = [1, 7, 9, 9, 8, 3]
s.nextPermutation(nums)
assert nums == [1, 8, 3, 7, 9, 9]

nums = [1, 2, 3]
s.nextPermutation(nums)
assert nums == [1, 3, 2]

nums = [5, 4, 3, 1, 2]
s.nextPermutation(nums)
assert nums == [5, 4, 3, 2, 1]
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        """ 
        3. Longest Substring Without Repeating Characters
        """
        # l, mx, hs = 0, 0, set()
        # for r in range(len(s)):
        #     while s[r] in hs:
        #         hs.remove(s[l])
        #         l += 1
        #     hs.add(s[r])
        #     mx = max(mx, r - l + 1)
        # return mx

        # l, mx, hs = 0, 0, set()
        # for r, v in enumerate(s):
        #     while v in hs:
        #         hs.remove(s[l])
        #         l += 1
        #     hs.add(v)
        #     mx = max(mx, r - l + 1)
        # return mx

        l, mx, hm = 0, 0, {}
        for i, v in enumerate(s):
            if v in hm:
                l = max(l, hm[v] + 1)
            hm[v] = i
            mx = max(mx, i - l + 1)
        return mx

solution = Solution()
assert solution.lengthOfLongestSubstring("abcabcbb") == 3
assert solution.lengthOfLongestSubstring("Hello World") == 6
```

```python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        """ 
        621. Task Scheduler
        """
        # Formula: (top_freq - 1) * (n + 1) + num_of_top_freq 
        counter = Counter(tasks).most_common()
        top_freq = counter[0][1]

        num_of_top_freq = 1
        for i in range(1, len(counter)):
            if counter[i][1] == counter[0][1]:
                num_of_top_freq += 1

        return max((top_freq - 1) * (n + 1) + num_of_top_freq, len(tasks))

s = Solution()
s.leastInterval(tasks = ["A","A","A","B","B","B", "C", "D", "E", "F"], n = 2)
```
